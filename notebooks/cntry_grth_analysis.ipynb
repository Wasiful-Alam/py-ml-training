```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def cntry_info_df():
    data = {
        'country':['Canada','England', 'France', 'Japan', 'Australia', 'Czech', 'India', 'Bangladesh', 'USA'],
        'language':['English','English','French', 'japanese', 'English', 'Czech', 'Hindi', 'Bangla', 'English'],
        'area':['60','57','30','25','36','26','65','12','55'],
        'population':['10', '18', '25', '28', '15', '16', '35', '50', '30' ],
        'GDP':['2','3.2','3.1','6','1.6','0.27','3.28','0.3','24',]
    }
    df=pd.DataFrame(data)
    return df
```

```python
cun_df= cntry_info_df()
cun_df
```

```python
def cntry_migrate_df():
    data = {
        'country':['Canada','England', 'France', 'Japan', 'Australia', 'Czech', 'India', 'Bangladesh', 'USA'],
        'migration':['0.6', '0.8', '1.4', '4', '1.2', '0.9','17', '7','0.2' ],
    }
    df=pd.DataFrame(data)
    return df
```

```python
mig_df = cntry_migrate_df()
mig_df
```

```python
merged_df = pd.merge(cun_df,mig_df,left_on='country', right_on = 'country')
merged_df
```
