```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.title('First graph of velocity',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})# naming the graph chart and using font types
plt.xlabel('time')
plt.ylabel('velocity')# defining axis

plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.title('First graph of velocity',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})# naming the graph chart and using font types
plt.xlabel('time')
plt.ylabel('velocity')# defining axis

plt.xticks([0,1,2,3])
plt.yticks([10,20,30]) #parameters in x and y axis are changed
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y,label='2x',color='red', linewidth=5, marker='.',markersize=20,markeredgecolor='blue',linestyle='--' )
plt.legend()#using the lable it changes the color of the line

plt.title('First graph of velocity',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('time')
plt.ylabel('velocity')

plt.xticks([0,1,2,3])
plt.yticks([10,20,30]) 
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y,'r.--',label='2x')#short notations can also be used
plt.legend()#using the lable it changes the color of the line

plt.title('First graph of velocity',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('time')
plt.ylabel('velocity')

plt.xticks([0,1,2,3])
plt.yticks([10,20,30]) 
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.title('First graph of velocity',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('time')
plt.ylabel('velocity')

x2= np.arange(0,4.5,0.5)
plt.plot(x2,x2**2)#creates a second curved line


plt.xticks([1,2,3])
plt.yticks([2,4,6,8]) 

plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.title('First graph of velocity',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('time')
plt.ylabel('velocity')

x2= np.arange(0,4.5,0.5)
plt.plot(x2[:4],x2[:4]**2)#shows first 4 values
plt.plot(x2[:6],x2[:6]**2,'--')#shows next 5 values in different way


plt.xticks([1,2,3])
plt.yticks([2,4,6,8]) 

plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]

plt.figure(figsize=(2,3),dpi=200)#               resizes the graph

plt.plot(x,y)
plt.title('First graph of velocity',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('time')
plt.ylabel('velocity')

x2= np.arange(0,4.5,0.5)
plt.plot(x2[:4],x2[:4]**2)
plt.plot(x2[:6],x2[:6]**2,'--')



plt.xticks([1,2,3])
plt.yticks([2,4,6,8]) 

plt.savefig('mygraph.png',dpi=300)#    saves the png of graph in current directory

plt.show()
```

```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values)

bars[0].set_hatch('/')# 0th bar is marked
bars[1].set_hatch('o')

plt.figure(figsize=(6,4))
plt.show()
```

```python
labels=['A','B','C','D']
values=[1,4,2,6]
bars=plt.bar(labels,values)

patterns=['/','-','|','.']
for bar in bars:
    bar.set_hatch(patterns.pop(0))#from second item to the rest are marked according to pattern list

plt.figure(figsize=(6,4))
plt.show()
```
