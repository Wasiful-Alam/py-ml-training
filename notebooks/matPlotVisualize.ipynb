```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y,'r-')
plt.xlabel('time')
plt.ylabel('velocity')
plt.title('speed of truck')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
 fig=plt.figure()#object orinted method
    
axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('time')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])#up down left right coordinates
```

```python
fig=plt.figure()

axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis2=fig.add_axes([0.2,0.5,0.4,0.3])#up down left right image coordinates

axis1.plot(x,y)
axis2.plot(y,x)

axis1.set_xlabel('time')
axis1.set_ylabel('velocity')
axis1.set_title('speed of truck')

axis2.set_xlabel('velocity')
axis2.set_ylabel('time')
axis2.set_title('speed of truck')
```

```python
fig,axis=plt.subplots(nrows=1,ncols=2) #iterating through
plt.tight_layout()
for current_ax in axis:
    current_ax.plot(x,y)
```

```python
fig,axis=plt.subplots(nrows=1,ncols=2) #calling using index
axis[0].plot(x,y)
axis[0].set_title('First Plot')

axis[1].plot(y,x)
axis[1].set_title('2nd Plot')
```

```python
fig,axes=plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig=plt.figure()#object orinted method
    
axes=fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')
axes.set_xlabel('time')

axes.legend(loc=0)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00', linewidth=20, alpha=0.2,
        linestyle='--',
        marker='*',markersize=30, markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')

ax.set_xlim([0,1])
ax.set_xlim([0,2])
```
